```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", 300)
```

```python
R1 = pd.read_csv('SLR_Area.csv')
R1
```

```python
R2 = pd.read_csv('SLR_Qual.csv')
R2
```

```python
combo = pd.concat([R1, R2], axis=1)
combo
```

```python
plt.style.use('default')
plt.style.use('ggplot')

plt.figure(figsize=(8,4))
sns.boxplot(data=combo[['SLR R2 Area', 'SLR R2 Qual']], showmeans=True,
                        meanprops={"marker":"o",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"black",
                      "markersize":"8"}, palette='coolwarm')
plt.xticks(rotation=25)
plt.ylabel('R-Squared', fontsize=14)
plt.xlabel('Model', fontsize=14)
plt.title('K-Fold CV R2 Comparison By Model', fontsize=18, color='blue')

plt.tight_layout();
```

```python
plt.style.use('default')
plt.style.use('ggplot')

plt.figure(figsize=(8,4))
sns.boxplot(data=combo[['SLR MAE Area', 'SLR MAE Qual']], showmeans=True,
                        meanprops={"marker":"o",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"black",
                      "markersize":"8"}, palette='coolwarm')
plt.xticks(rotation=25)
plt.ylabel('Mean Absolute Error (US$)', fontsize=14)
plt.xlabel('Model', fontsize=14)
plt.title('K-Fold CV MAE Comparison By Model', fontsize=18, color='blue')

plt.tight_layout();
```

```python
combo.describe()
```
