## Ridge ##

```python
model = Ridge()
scaler = preprocessing.StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
alphas = 10**np.linspace(10,-2,100)*0.5
ridgecv = RidgeCV(alphas = alphas, scoring = 'neg_mean_squared_error')
ridgecv.fit(X_train_scaled, y_train)
ridgecv.alpha_
```

```python
ridge_best_alpha = Ridge(alpha=ridgecv.alpha_)
ridge_best_alpha.fit(X_train_scaled,y_train)
```

```python
ridge_best_alpha.score(X_train_scaled,y_train)
```

```python
#R2 for test set
ridge_best_alpha.score(X_test_scaled, y_test)
```

## Lasso ##

```python
lassocv = LassoCV(alphas = alphas, cv = 10, max_iter = 100000)
lassocv.fit(X_train_scaled, y_train)
lassocv.alpha_
```

```python
lasso_best_alpha = Lasso(alpha=lassocv.alpha_)
lasso_best_alpha.fit(X_train_scaled,y_train)
```

```python
lasso_best_alpha.score(X_train_scaled, y_train)
```

```python
#R2 for test set
lasso_best_alpha.score(X_test_scaled, y_test)
```
