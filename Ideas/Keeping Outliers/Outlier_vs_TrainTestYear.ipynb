```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()
```

```python
train = pd.read_csv('../Data/TrainYear.csv')
```

```python
train.shape
```

```python
train = train.assign(TrainTest='Train')
```

```python
train.shape
```

```python
test = pd.read_csv('../Data/TestYear.csv')
```

```python
test.shape
```

```python
test = test.assign(TrainTest='Test')
```

```python
test.shape
```

```python
Combo = pd.concat([train,test]).reset_index(drop=True)
Combo.shape
```

```python
sns.mpl.rcParams['figure.figsize'] = (10.0, 6.0)

sns.scatterplot(data=Combo, x='GrLivArea', y='SalePrice', 
                hue='TrainTest', palette='deep');
```
