### If NaN implies "None" or "No Feature", replace it with "None": ###

```python
#Create Copy of dataframe and 'None' Impute for selected features: 

c_new = c.copy()

for column in [
    'PoolQC',
    'MiscFeature',
    'Alley',
    'Fence',
    'FireplaceQu',
    'GarageFinish',
    'GarageCond',
    'GarageQual',
    'GarageType',
    'BsmtExposure',
    'BsmtFinType2',
    'BsmtFinType1',
    'BsmtCond',
    'BsmtQual',
    'MasVnrType'
]:
    
    c_new[column] = c[column].fillna('None')
        
```

### If NaN doesn't imply "None" or have a meaning, replace it with "Mode": 

```python
c.Electrical.value_counts(dropna=False)
#Electrical: Doesn't have a designated meaning for "NA", so we will 'mode' impute: 
```

```python
#Electrical: Only feature to "mode" impute
c_new['Electrical'] = c['Electrical'].fillna(c['Electrical'].mode()[0])
```
