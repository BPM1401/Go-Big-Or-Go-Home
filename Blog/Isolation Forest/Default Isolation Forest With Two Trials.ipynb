## Default Isolation Forest With Two Trials

```python
from sklearn.ensemble import IsolationForest
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", 300)
import seaborn as sns
```

```python
ordinal = pd.read_csv('../Data/BaseOrdinal.csv')
ordinal.shape
```

```python
ordinal = ordinal.drop(['Unnamed: 0'], axis=1)
ordinal.shape
```

```python
X = ordinal.values
#y = ordinal.SalePrice.values
```

```python
preds = []
decs = []

for i in range(2):
    #iso = IsolationForest(n_estimators=1000, max_samples=750, contamination=0.01).fit(X)
    iso = IsolationForest().fit(X)
    pred = pd.Series(iso.predict(X))
    dec = pd.Series(iso.decision_function(X))
    preds.append(pred)
    decs.append(dec)
    
    #result_pred = pd.concat([pred], axis=1)
    #result_dec = pd.concat([dec], axis=1)
    
```

```python
combined_pred = pd.concat(preds, axis=1)
combined_pred
```

```python
combined_pred['Pred_Sum'] = combined_pred.sum(axis=1)
combined_pred
```

```python
combined_dec = pd.concat(decs, axis=1)
combined_dec
```

```python
combined_dec['Dec_Sum'] = combined_dec.sum(axis=1)
combined_dec
```

```python
final = combined_pred.join(combined_dec, lsuffix='pred', rsuffix='dec')
final.sort_values(by='Dec_Sum')
```
