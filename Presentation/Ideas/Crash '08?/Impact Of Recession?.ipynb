```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 100)
```

```python
df = pd.read_csv('../../Data/Original.csv')
```

```python
df = df[['YrSold', 'MoSold', 'SalePrice']].rename(columns={'YrSold':'Year', 'MoSold':'Month'})
df
```

```python
df['Date'] = pd.to_datetime(df[['Year', 'Month']].assign(DAY=1))
```

```python
df
```

```python
plt.style.use('default')
plt.style.use('ggplot')

plt.figure(figsize=(8,4))

plt.title('House Price Transactions Grouped By Year (Ames, Iowa)', fontsize=18, color='blue')

sns.boxplot(data=df, x='Year', y='SalePrice', showmeans=True,
                        meanprops={"marker":"o",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"black",
                      "markersize":"8"})

plt.ylabel('House Sale Price (US$)', fontsize=14)
plt.xlabel('Year', fontsize=14)

plt.tight_layout();
```
