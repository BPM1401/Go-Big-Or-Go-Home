### Create Full (Clean) + Final Feature Engineering ###

```python
full = pd.concat([c, n], axis=1)
full.shape
```

```python
full.isnull().sum().sum()
```

```python
full.describe()
```

```python
full.head(3)
```

```python
full['QA'] = full['GrLivArea'] * full['OverallQual']
```

```python
full
```

```python
full['MoSold']
```

```python
-np.cos(0.5236 * full['MoSold'])
```

```python
full['MoSold'] = -np.cos(0.5236 * full['MoSold'])
```

```python
full['SqFtPerRoom'] = full['GrLivArea'] / (full['TotRmsAbvGrd'] + 
                                           full['FullBath'] +
                                           full['HalfBath'] +
                                           full['KitchenAbvGr']) 

full['Total_Hm_Qual'] = full['OverallQual'] + full['OverallCond']

full['Total_Bathrooms'] = full['FullBath'] + (0.5 * full['HalfBath'] +
                                              full['BsmtFullBath'] + 
                                              0.5 * full['BsmtHalfBath'])

full['HighQualSF'] = full['1stFlrSF'] + full['2ndFlrSF']
```
