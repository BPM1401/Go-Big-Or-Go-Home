### Convert Supposingly Numerical Cols To Categorical ###

```python
df[['MSSubClass','YrSold','MoSold']].isnull().sum()
```

```python
df['MSSubClass'] = df['MSSubClass'].apply(str)
df['YrSold'] = df['YrSold'].astype(str)
df['MoSold'] = df['MoSold'].astype(str)
```

### Object Columns, Fillna with "None" & Dummify! ###

```python
c = df.select_dtypes('object')
c.head(3)
```

```python
c.shape
```

```python
c.isnull().sum().sum()
```

```python
c = c.fillna("None")
c.isnull().sum().sum()
```

```python
c.shape
```

```python
c = pd.get_dummies(c, drop_first=True)
c.shape
```

```python
c
```

### Numeric Columns, Fillna with Mean ###

```python
n = df.select_dtypes('number')
n.head(3)
```

```python
n.isnull().sum()
```

```python
nan_cols = [i for i in n.columns if n[i].isnull().any()]
nan_cols
```

```python
n[nan_cols].isna().sum()
```

```python
Rows_w_Nan = n[n[nan_cols].isna().any(axis=1)]
Index = Rows_w_Nan.index
Rows_w_Nan
```

```python
NanDF = Rows_w_Nan[nan_cols]
NanDF
```

```python
n[nan_cols].describe()
```

```python
n[nan_cols].mean()
```

```python
for cols in nan_cols:
    mean_value = n[cols].mean()
    n[cols].fillna(value=mean_value, inplace=True)
    print (cols, mean_value)
```

```python
#All NaN's have been replaced by Column Mean Values. Can double check below against Rows_w_Nan Dataframe: 
n.loc[Index]
```
