```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()

import scipy as sp
from sklearn.linear_model import LinearRegression

pd.set_option("display.max_columns", 100)

import statsmodels.api as sm
import statsmodels.stats.api as sms
from scipy import stats
```

```python
df = pd.read_csv('Ames_HousePrice.csv')
```

### Object Columns, Fillna with "None" & Dummify! ###

```python
c = df.select_dtypes('object')
c.head(3)
```

```python
c.shape
```

```python
c.isnull().sum().sum()
```

```python
c = c.fillna("None")
```

```python
c.isnull().sum().sum()
```

```python
c.shape
```

```python
c = pd.get_dummies(c, drop_first=True)
c.shape
```

```python
c
```

### Numeric Columns, Dropna on Rows/Observations, Create Full (Clean) ###

```python
n = df.select_dtypes('number')
n.head(3)
```

```python
n.shape
```

```python
n.isnull().sum().sum()
```

```python
full = pd.concat([c, n], axis=1)
full.shape
```

```python
full = full.dropna()
full.shape
```

```python
full.isnull().sum().sum()
```

### MLR For Base Case ###

```python
y=full.SalePrice
X=full.drop('SalePrice', axis=1)
```

```python
regressor = LinearRegression()
regressor.fit(X,y)
```

```python
regressor.score(X,y)
```

```python
#Root Mean Squared Error for entire set:
np.sqrt(np.mean((regressor.predict(X) - y)**2))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30)
```

```python
#Root Mean Squared Error for "train" set:
rmse_train = np.sqrt(np.mean((regressor.predict(X_train) - y_train)**2))
rmse_train
```

```python
#Root Mean Squared Error for "test" set:
rmse_test = np.sqrt(np.mean((regressor.predict(X_test) - y_test)**2))
rmse_test
```

```python
ols = LinearRegression()
ols.fit(X_train, y_train)
print("R^2 for train set: %f" %ols.score(X_train, y_train))
print("Root Mean Squared Error for train set: %f" %rmse_train)

print('-'*50)

print("R^2 for test  set: %f" %ols.score(X_test, y_test))
print("Root Mean Squared Error for test set: %f" %rmse_test)
```

### Stats for MLR For Base Case ###
