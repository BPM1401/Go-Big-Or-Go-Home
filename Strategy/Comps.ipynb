```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", 300)
```

```python
R1 = pd.read_csv('Linear.csv')
R1
```

```python
R2 = pd.read_csv('Non-Linear.csv')
R2
```

```python
combo = pd.concat([R1, R2], axis=1)
combo
```

```python
sns.boxplot(data=combo[['Ridge R2', 'Lasso R2', 'Elastic R2', 
                        'GB R2', 'XGB R2', 'Cat R2']], showmeans=True,
                        meanprops={"marker":"o",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"black",
                      "markersize":"8"})
plt.xticks(rotation=25);
```
