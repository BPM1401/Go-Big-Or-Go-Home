### Object Columns, Fillna with "None" & Dummify! ###

```python
c = df.select_dtypes('object')
c.head(3)
```

```python
c.shape
```

```python
c.isnull().sum().sum()
```

```python
c = c.fillna("None")
```

```python
c.isnull().sum().sum()
```

```python
c.shape
```

```python
c = pd.get_dummies(c, drop_first=True)
c.shape
```

```python
c
```

### Numeric Columns, Dropna on Rows/Observations, Create Full (Clean) ###

```python
n = df.select_dtypes('number')
n.head(3)
```

```python
n.shape
```

```python
n.isnull().sum().sum()
```

```python
full = pd.concat([c, n], axis=1)
full.shape
```

```python
full = full.dropna()
full.shape
```

```python
full.isnull().sum().sum()
```

### Ridge For Base Case (Alpha=0 is same as MLR Base) ###

```python
y=full.SalePrice
X=full.drop('SalePrice', axis=1)
```

```python
##Note, Alpha = 0 is basically the same as OLS / LinearRegression()

ridge = Ridge(alpha=0.00001)
ridge.fit(X,y)
```

```python
ridge.score(X,y)
```

```python
#Root Mean Squared Error for entire set:
np.sqrt(np.mean((ridge.predict(X) - y)**2))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30)
```

```python
#Root Mean Squared Error for "train" set:
rmse_train = np.sqrt(np.mean((ridge.predict(X_train) - y_train)**2))
rmse_train
```

```python
#Root Mean Squared Error for "test" set:
rmse_test = np.sqrt(np.mean((ridge.predict(X_test) - y_test)**2))
rmse_test
```

```python
ridge.fit(X_train, y_train)
print("R^2 for train set: %f" %ridge.score(X_train, y_train))
print("Root Mean Squared Error for train set: %f" %rmse_train)

print('-'*50)

print("R^2 for test  set: %f" %ridge.score(X_test, y_test))
print("Root Mean Squared Error for test set: %f" %rmse_test)
```

### Must SCALE inputs before Regularization! ###

##### Standardizing Parameters (Mean & STD) Must Be Learnt From Training Data #####

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30)
```

```python
from sklearn import preprocessing

# Fit Ridge on Standardized data
ridge_alpha = Ridge()
scaler = preprocessing.StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

#ridge.fit(X_train_scaled , y_train)
#scaled_predictions = ridge.predict(X_test_scaled)
```

```python
alphas = 10**np.linspace(10,-2,100)*0.5
alphas
```

```python
#ridge = Ridge(normalize = True)
coefs = []

for a in alphas:
    ridge_alpha.set_params(alpha = a)
    ridge_alpha.fit(X_train_scaled, y_train)
    coefs.append(ridge_alpha.coef_)
    
np.shape(coefs)
```

```python
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
```

```python
from sklearn.linear_model import RidgeCV
from sklearn.metrics import mean_squared_error

ridgecv = RidgeCV(alphas = alphas, scoring = 'neg_mean_squared_error')
ridgecv.fit(X_train_scaled, y_train)
ridgecv.alpha_
```

```python
ridge_best_alpha = Ridge(alpha=ridgecv.alpha_)
ridge_best_alpha.fit(X_train_scaled,y_train)
```

```python
ridge_best_alpha.score(X_train_scaled,y_train)
```

```python
#Root Mean Squared Error for entire set:
np.sqrt(np.mean((ridge_best_alpha.predict(X_train_scaled) 
                 - y_train)**2))
```

```python
#R2 for test set
ridge_best_alpha.score(X_test_scaled, y_test)
```

```python
#Root Mean Squared Error for test set:
np.sqrt(np.mean((ridge_best_alpha.predict(X_test_scaled) 
                 - y_test)**2))
```
